```python
plt.plot(df1['Sepal_Length'] , df1['Sepal_Width'], ls = '' , marker = 'o')
```

```python
plt.plot(df1.Sepal_Length , df1.Sepal_Width , ls= '' , marker = 'o' , label='Sepal')
plt.plot(df1.Petal_Length , df1.Petal_Width , ls= '' , marker = 'o' , label = 'Petal')
plt.legend()
plt.show()
```

```python

plt.hist(df1.Petal_Length , bins=25)
```

```python
fig , ax = plt.subplots()
print(type(ax))
ax.barh(np.arange(10),df1['Sepal_Width'].iloc[:10])

ax.set_yticks(np.arange(0.01,10,1.0,))
ax.set_yticklabels(np.arange(1,11))
ax.set(xlabel='xlabel' , ylabel='ylabel' , title = 'Title')
```
